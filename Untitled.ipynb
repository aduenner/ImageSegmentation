```python
import boto3
import boto3.s3
import sys
import glob
```

```python
s3 = boto3.client('s3')
```

```python
bucket_name = 'duennerstore'
```

```python
filenames = glob.glob('/home/andrew/ImageSegmentation/Training/Raw_Images_1024/*.png')
for file in filenames:
    split_path = file.split('/')
    new_path = 'PNG_1024/'+split_path[-1]
    s3.upload_file(file, bucket_name, new_path)
    print(new_path)
```
