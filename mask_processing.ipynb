```python
import sys
import os
sys.path.insert(0, os.getcwd()+"/Utils")
from skimage.io import imread, imsave, imshow
import numpy as np
import glob
import matplotlib.pyplot as plt
import mask_functions as mfun
```

```python
import os
names = glob.glob('*png')

img_test = names[12]
img = imread(img_test)
imshow(img)
plt.title('Original Image')
plt.show()


colors = mfun.original_color_dict()

mask_dict = mfun.binarize_mask(img_test,colors)


for i,mask in enumerate(mask_dict):
    plt.subplot(2,2,i+1)
    plt.imshow(mask_dict[mask])
    plt.title('mask ='+mask)
plt.show()


img = mfun.colorize_mask_dict(mask_dict,colors)
plt.imshow(img)
plt.title('recolored image')
plt.show()
```
